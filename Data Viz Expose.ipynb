```python
##  Pie Charts
```

```python
from matplotlib import pyplot as plt
```

```python
slices=[7,2,2,13]
activities=['sleeping','eating','working','playing']
cols=['c','m','r','b']

plt.pie(slices,
        labels=activities,
        colors=cols,
        startangle=90,
        shadow=True,
        explode=(0,0.1,0,0.1),
        autopct='%1.1f%%')
        
plt.title('Pie Chart')
 
```

```python
import numpy as np
```

```python
def f(t): 
    return np.exp(-t) * np.cos(2*np.pi*t)

t1=np.arange(0.0,5.0,0.1)
t2=np.arange(0.0,5.0,0.02)

plt.subplot(211)
plt.plot(t1,f(t1),'bo',t2,f(t2))

plt.subplot(212)
plt.plot(t2,np.cos(2*np.pi*t2))
plt.show()
```
